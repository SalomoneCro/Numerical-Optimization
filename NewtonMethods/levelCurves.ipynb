```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the functions based on the image
def f1(x1, x2):
    return 4 * x1**2 - 20 * x1 + (1/4) * x2**2 + 8

def f2(x1, x2):
    return (1/2) * x1 * x2**2 + 2 * x1 - 5 * x2 + 8

# Create a grid of values
x1 = np.linspace(-10, 10, 400)
x2 = np.linspace(-10, 10, 400)
X1, X2 = np.meshgrid(x1, x2)

# Evaluate the functions over the grid
F1 = f1(X1, X2)
F2 = f2(X1, X2)

# Plot the level curves
plt.figure(figsize=(8, 6))
contour1 = plt.contour(X1, X2, F1, levels=[0], colors='blue', linewidths=2)
contour2 = plt.contour(X1, X2, F2, levels=[0], colors='red', linewidths=2)

# Adding labels and titles
plt.title("Level Curves of the Functions")
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(True)

# Show plot
plt.show()
```

```python
# Define the new functions from the second image
def ff1(x1, x2):
    return np.sin(4 * np.pi * x1 * x2) - 2 * x2 - x1

def ff2(x1, x2):
    return (4 * np.pi - 1) / (4 * np.pi) * (np.exp(2 * x1) - np.e) + 4 * np.e * x2**2 - 2 * np.e * x1

# Create a grid of values
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(x1, x2)

# Evaluate the functions over the grid
F1 = ff1(X1, X2)
F2 = ff2(X1, X2)

# Plot the level curves
plt.figure(figsize=(8, 6))
contour1 = plt.contour(X1, X2, F1, levels=[0], colors='blue', linewidths=2, linestyles='solid')
contour2 = plt.contour(X1, X2, F2, levels=[0], colors='red', linewidths=2, linestyles='solid')

# Adding labels and titles
plt.title("Level Curves of the Functions")
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.grid(True)

# Show plot
plt.show()
```
